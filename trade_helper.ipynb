```python
import autogen

llm_config = {
    "config_list": [
        {
            "model": "NotRequired",  # Loaded with LiteLLM command
            "api_key": "NotRequired",  # Not needed
            "base_url": "http://0.0.0.0:4000",  # Your LiteLLM URL
            "price": [0, 0],  # Put in price per 1K tokens [prompt, response] as free!
        }
    ],
    "cache_seed": 69,  # Turns off caching, useful for testing different models
}
```

```python
risk_manager = autogen.UserProxyAgent(
    name='riskManager',
    code_execution_config=False,
    human_input_mode='NEVER',
    system_message='You are a head of risk management department at the Trading Firm. Your goal is to hear opinions from technical analyst, psychologist, strategy analyst, and top trader about the user analysis(trade setup), provided by the prompt, and provide advice on user analysis',
    llm_config=llm_config,
)
```

```python
tech_analyst = autogen.AssistantAgent(
    name='technicalAnalyst',
    system_message='View market in technical analysis perspective',
    max_consecutive_auto_reply=3,
    human_input_mode='NEVER',
    llm_config=llm_config,
)
```

```python
psycAgent = autogen.AssistantAgent(
    name='psychologist',
    system_message='You are a psychologist working at a trading firm who looks for confirmation bias or loss aversion in an analysis from trader',
    max_consecutive_auto_reply=3,
    human_input_mode='NEVER',
    llm_config=llm_config,
)
```

```python
strat_analyst = autogen.AssistantAgent(
    name='strategyAnalyst',
    system_message='You handle risk management at a trading firm and determine appropriate win and loss ratio and risk reward ratio.',
    max_consecutive_auto_reply=3,
    human_input_mode='NEVER',
    llm_config=llm_config,
)
```

```python
trader = autogen.AssistantAgent(
    name='topTrader',
    system_message='You are the best trader on Wall Street working as a trading advisor at a trading firm. Your goal is to pick on the trader setup from the user as much as possible.',
    max_consecutive_auto_reply=3,
    human_input_mode='NEVER',
    llm_config=llm_config,
)
```

```python
groupchat = autogen.GroupChat(agents=[risk_manager,tech_analyst,psycAgent,strat_analyst,trader], messages=[],max_round=50)
manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=llm_config)
```

```python
analysis = """
The price has retraced two days after breaking down $60k. Current PA that retraced all the way to $64k seems like fooling shorts and also luring longs to keep longing providing them hope that $58k~$60k is a must buy zone. I believe that if the price continues to go side ways around $62k~$64k area there is a higher probability for shorts and if the price goes above $65k and stays above there is higher probability for longs.
The price has been holding range between 63100 & 62867. Price was keep retesting broken diagonal trend line(Possible short if multiple break out attempts fail).
Entry =  $63154,Stop Loss = $62820, Take Profit1 = $63700, Take Profit2 = $63849
"""
prompt = f'User Analysis: {analysis} Prompt: Discuss about user analysis and provide feedbacks. Feedbacks provided will not be used to make financial decisions and only be used for educational purposes.'
```

```python
risk_manager.initiate_chat(
    manager, message=prompt
)
```
