```python
import autogen

llm_config = {
    "config_list": [
        {
            "model": "NotRequired",  # Loaded with LiteLLM command
            "api_key": "NotRequired",  # Not needed
            "base_url": "http://0.0.0.0:4000",  # Your LiteLLM URL
            "price": [0, 0],  # Put in price per 1K tokens [prompt, response] as free!
        }
    ],
    "cache_seed": 69,  # Turns off caching, useful for testing different models
}
```

```python
moderator = autogen.UserProxyAgent(
    name = 'Moderator',
    code_execution_config=False,
    human_input_mode="NEVER",
    system_message="You are a moderator to the conversation about Presidential election between Donald Trump and Kamala Harris in 2024",
    llm_config=llm_config
)
```

```python
rep_support = autogen.AssistantAgent(
    name='Republican_Supporter',
    system_message='You are a Republican supporter who thinks that Donald Trump will win the 2024 presidential election',
    max_consecutive_auto_reply=5,
    human_input_mode='NEVER',
    llm_config=llm_config

)

dem_support = autogen.AssistantAgent(
    name='Democrat_Supporter',
    system_message='You are a Democrat supporter who thinks that Kamala Harris will win the 2024 presidential election',
    max_consecutive_auto_reply=5,
    human_input_mode='NEVER',
    llm_config=llm_config
)
```

```python
groupchat = autogen.GroupChat(agents=[moderator, rep_support, dem_support], messages=[], max_round=30)
manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=llm_config)
```

```python
moderator.initiate_chat(
    manager, message="Between Kamala Harris and Donald Trump, Who would win the presidential election in 2024?"
)
```
